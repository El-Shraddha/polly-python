### Querying for dataset-id, description, number of samples etc. from GEO where platform is 'RNASeq', number of samples >20.

```python
query = """SELECT dataset_id, description, total_num_samples, disease, tissue FROM geo.datasets 
            WHERE platform = 'RNASeq' AND total_num_samples > 20 ORDER BY total_num_samples desc"""
results=omixatlas.query_metadata(query, query_api_version="v2")
results
```

### Querying for datasets from GEO where disease array has any item consisting of keyword 'neoplasm'

```python
#Querying for datasets from GEO where disease array has any item consisting of keyword 'neoplasm'
query = "SELECT disease FROM geo.datasets WHERE CARDINALITY(FILTER(disease, d -> LOWER(d) like '%neoplasm%')) > 0"
results=omixatlas.query_metadata(query, query_api_version="v2")
results
```

### Get number of datasets with all combinations of disease and tissues in GEO

## JOIN Queries

### Get microarray datasets and their samples where the ‘HeLa’ cell line was used on a microarray platform

```python
#Get microarray datasets and their samples where the ‘HeLa’ cell line was used on a microarray platform

query = """  SELECT datasets.dataset_id AS dataset_id, sample_name AS sample_name FROM geo.datasets AS datasets
    JOIN (
        SELECT src_dataset_id AS dataset_id, name AS sample_name FROM geo.samples WHERE CONTAINS(cell_line, 'HeLa')
         ) AS hela_samples
    ON datasets.dataset_id = hela_samples.dataset_id WHERE LOWER(platform) = 'microarray'
    ORDER BY datasets.dataset_id"""
results=omixatlas.query_metadata(query, query_api_version="v2")
results
```

### Queries using a sub-query

```python
query = """SELECT src_dataset_id AS dataset_id, name AS sample_name FROM geo.samples
    WHERE CONTAINS (cell_line, 'HeLa') AND src_dataset_id 
    IN (SELECT dataset_id FROM geo.datasets WHERE LOWER (platform) = 'microarray')"""
results=omixatlas.query_metadata(query, query_api_version="v2")
results
```

### Parallelise queries

```python
def query_split(query):
    
    """
    Returns N SQL queries equal to the number of input indexes
    
    Extended description of function.
    
    Parameters:
    query (str): input SQL query to search across any given omixatlases
    
    Returns:
    q_list (list): list of SQL queries where each query queries on a single index only.
    
    """
    
    index_search = re.search("from(.*)where", query, re.IGNORECASE)
    index_list = index_search.group(1).split(',')
    index_list = [index.strip() for index in index_list]
    
    const_string1 = query.split(index_list[0]) #Extract constant SQL query
    const_string2 = query.split(index_list[-1]) #Extract constant SQL query
    
    q_list = []
    
    for index in index_list:
        q = f"{const_string1[0]}{index}{const_string2[-1]}"
        q_list.append(q)
    
    return q_list

def query_oa(query):
    
    """
    Returns a table containing metadata for all results based on the input SQL query
    
    Extended description of function.
    
    Parameters: 
    query (string): list of input SQL queries where each query is used to query across a single omixatlas
    
    Returns:
    all_df (dataframe): DataFrame containing metadata of the resulting data.
    
    """

    all_df = omixatlas.query_metadata(query)
    return all_df
    

def parallel_query(query_list):
    
    """
    Performs a parallel execution by dividing the task among multiple threads rather than multiple CPUs
    
    Extended description of function.
    
    Parameters: 
    query_list (list): list of input SQL queries where each query is used to search across a single omixatlas
    
    Returns:
    result_df (dataframe): DataFrame containing dataset level metadata of the resulting datasets.
    
    """

    final = []
    result = Parallel(n_jobs=10, prefer="threads")(delayed(query_oa)(query) for query in query_list)

    for that_dict in result:
        if isinstance(that_dict, pd.DataFrame):
            final.append(that_dict)
    
    result_df = pd.concat(final)
    
    return result_df
```
